```python
#import pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#get data
Arrest_Data = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f")

#only necessary columns
fixed_arrest_data = Arrest_Data[['OFFENSES','INCIDENTNEIGHBORHOOD']]

#added up arrests in each neighborhood
count = fixed_arrest_data.groupby('INCIDENTNEIGHBORHOOD').count()
print(count)

#plot all points
count['OFFENSES'].plot(kind='bar', figsize=(12, 8), color='skyblue', edgecolor='black')
plt.show()
```

```python
#show lowest points
filtered_count = count[count['OFFENSES']<100]
print(filtered_count)
filtered_count['OFFENSES'].plot(kind='bar', figsize=(12, 8), color='skyblue', edgecolor='black')
plt.show()
```
