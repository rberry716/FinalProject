```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fire.csv")
```

```python
plot = df["neighborhood"].value_counts().sort_index().plot.bar(y='neighborhood', 
                                                  subplots=False, 
                                                  figsize=(20,10 ))
```

```python

filtered= df["neighborhood"].value_counts()
filtered= filtered[filtered < 50].sort_index()

filtered.plot.bar(figsize=(20, 10))
plt.xlabel("Neighborhood")
plt.ylabel("Number of Fire Reports")
plt.title("Neighborhood Fires Less Than 50")
plt.show()
```
