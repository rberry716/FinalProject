```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
citations = pd.read_csv("Non-TrafficCitations.tsv", sep = "\t")
```

```python
citationsNeighborhoods = citations["NEIGHBORHOOD"].value_counts().reset_index()
citationsNeighborhoods.columns = ["NEIGHBORHOOD", "COUNT"]
citationsNeighborhoods = citationsNeighborhoods[citationsNeighborhoods["NEIGHBORHOOD"]!= "Outside County"]
citationsNeighborhoods = citationsNeighborhoods[citationsNeighborhoods["NEIGHBORHOOD"]!= "Outside State"]
citationsNeighborhoods = citationsNeighborhoods[citationsNeighborhoods["NEIGHBORHOOD"]!= "Outside City"]
citationsNeighborhoods = citationsNeighborhoods[citationsNeighborhoods["NEIGHBORHOOD"]!= "Unable To Retrieve Address"]
```

```python
citationsNeighborhoods.iloc[50: ].plot(kind="bar",x="NEIGHBORHOOD", y="COUNT")
plt.show()
```
